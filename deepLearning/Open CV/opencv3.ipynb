```python
import cv2

frame=cv2.imread('C:\\Users\\hp\\Desktop\\cv4.jpg')

cv2.imshow("my image",frame)
cv2.waitKey()
```

```python
from matplotlib import  pyplot as plt
%matplotlib inline
g_img=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)  #cv2.cvtColor is a function which convert BGR2RGB
plt.imshow(g_img)
plt.axis("off")
plt.show()
```

```python
haar_face_cascade=cv2.CascadeClassifier("C:\\Users\\hp\\Downloads\\face1.xml")
faces=haar_face_cascade.detectMultiScale(g_img,scaleFactor=1.06,minNeighbors=5);

        
for(x,y,w,h) in faces:
            cv2.rectangle(g_img,(x,y),(x+w,y+h),(0,255,0),2)
            color1=g_img[y:y+h,x:x+w]
            
            plt.imshow(color1)
            plt.axis('off')
            plt.show()

plt.imshow(g_img)
plt.axis('off')
plt.show()      
```
