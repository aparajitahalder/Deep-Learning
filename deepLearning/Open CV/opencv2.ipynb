```python
import cv2

frame=cv2.imread('C:\\Users\\hp\\Desktop\\cv4.jpg')

cv2.imshow("my image",frame)
cv2.waitKey()
```

```python
from matplotlib import  pyplot as plt
%matplotlib inline
plt.imshow(frame)
plt.axis("off")
plt.show()   #it take BGR
```

```python
g_img=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)  #cv2.cvtColor is a function which convert BGR2RGB
plt.imshow(g_img)
plt.axis("off")
plt.show()
```

```python
haar_face_cascade=cv2.CascadeClassifier("C:\\Users\\hp\\Downloads\\face1.xml")
#load cascade classifier training file for haarcascade #model with various face features
```

```python
faces=haar_face_cascade.detectMultiScale(g_img,scaleFactor=1.06,minNeighbors=5);
print("face found",len(faces))
```

```python
for(x,y,w,h) in faces:  #four cordinates of face
    cv2.rectangle(g_img,(x,y),(x+w,y+h),(0,255,0),2)
plt.imshow(g_img)
plt.axis('off')
plt.show()
```

# face detection in live video

```python
from IPython.display import clear_output

webcam=cv2.VideoCapture(0)
try:
    while True:
        _,frame=webcam.read()
        frame=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        plt.axis("off")
        plt.title("")    
        haar_face_cascade=cv2.CascadeClassifier("C:\\Users\\hp\\Downloads\\face1.xml")
        faces=haar_face_cascade.detectMultiScale(frame,scaleFactor=1.06,minNeighbors=5);
        
        for(x,y,w,h) in faces:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
       
        plt.imshow(frame,cmap="Greys_r")
        plt.show()
        clear_output(wait=True)
except KeyboardInterrupt:
    print("video Interrupt")
    webcam.release()        
```
