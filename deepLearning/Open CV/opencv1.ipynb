```python
#capture image from live video
video=cv2.VideoCapture(0)
count=0
while True:
        ret,frame=video.read()
        cv2.imshow('live video',frame)
        count+=1
        if cv2.waitKey(20)&count%20==0:
            frame_RGB=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
            plt.imshow(frame_RGB)
            cv2.imwrite('img'+str(count)+'.png',frame_RGB)  #will save in current folder
        
        elif cv2.waitKey(20)&0xFF==27:
             break
video.release()
cv2.destroyAllWindows()
```

# start and save live video

```python


webcam=cv2.VideoCapture(0)
cv2.namedWindow("s",cv2.WINDOW_AUTOSIZE)
#create videowriter object

fourcc=cv2.VideoWriter_fourcc(*'XVID')  #this function write the video in different formats
video=cv2.VideoWriter('sam.mp4',fourcc,20.0,(640,480))  #save the video
while webcam.isOpened():
    ret,frame=webcam.read()
    video.write(frame)
    cv2.imshow('s',frame)
    if cv2.waitKey(20)&0xFF==27:
        break
webcam.release()
video.release()
cv2.destroyAllWindows()
```

# drawing shapes in live video

```python
import cv2
webcam=cv2.VideoCapture(0)
cv2.namedWindow("drawing",cv2.WINDOW_AUTOSIZE)
msg=""

while webcam.isOpened():
    ret,frame=webcam.read()
         #(left top corner) (right bottom corner) (color)(width)
    cv2.rectangle(frame,(100,100),(530,450),(150,150,0),3)
    cv2.circle(frame,(280,170),20,(150,150,0),3)
    cv2.circle(frame,(350,170),20,(150,150,0),3)
    cv2.line(frame,(330,170),(300,170),(255,0,0),5)
    cv2.putText(frame,msg,(120,120),cv2.FONT_ITALIC,2,(150,150,0),2)
    cv2.imshow('drawing',frame)
    key=cv2.waitKey(100) & 0xFF
    if key not in [255,27]:
        msg+=chr(key)
    elif key==27:
        break
webcam.release()
cv2.destroyAllWindows()
```
