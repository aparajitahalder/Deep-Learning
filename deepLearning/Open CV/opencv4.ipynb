```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


haar_eye_cascade=cv2.CascadeClassifier('C:\\Users\\hp\\downloads\\haarcascade_eye.xml')
video=cv2.VideoCapture(0)
ret,frame=video.read()
frame_gray=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
eyes=haar_eye_cascade.detectMultiScale(frame_gray,1.3,5)
print(len(eyes))
for(x,y,w,h) in eyes:
     cv2.rectangle(frame_gray,(x,y),(x+w,y+h),(100,150,120),2)
plt.imshow(frame_gray)        
```

```python
import cv2

frame=cv2.imread('C:\\Users\\hp\\Desktop\\cv7.jpg')

cv2.imshow("my image",frame)
cv2.waitKey()
```

```python
from matplotlib import  pyplot as plt
%matplotlib inline
img=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)  #cv2.cvtColor is a function which convert BGR2RGB
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
import cv2
haar_face_cascade=cv2.CascadeClassifier("C:\\Users\\hp\\Downloads\\face1.xml")
faces1=haar_face_cascade.detectMultiScale(img,scaleFactor=1.06,minNeighbors=5);

x1=y1=0
s1=s2=1
for(x,y,w,h) in faces1:
            #cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
            color2=img[y:y+h,x:x+w]
            print(x,y,w,h)
            if h*w>s1*s2:
                s1=h
                s2=w
                x1=x
                y1=y
            
                #largest_face=img[y:y+h,x:x+w]
            plt.imshow(color2)
            plt.axis('off')
            plt.show()



cv2.rectangle(img,(x1,y1),(x1+s2,y1+s1),(0,255,0),2)
print("largest face")
plt.imshow(img)
plt.axis('off')
plt.show() 
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


haar_eye_cascade=cv2.CascadeClassifier('C:\\Users\\hp\\downloads\\haarcascade_eye.xml')
haar_face_cascade=cv2.CascadeClassifier("C:\\Users\\hp\\Downloads\\face1.xml")


video=cv2.VideoCapture(0)
ret,frame=video.read()
frame_gray=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
eyes=haar_eye_cascade.detectMultiScale(frame_gray,1.3,5)
faces=haar_face_cascade.detectMultiScale(frame_gray,scaleFactor=1.06,minNeighbors=5);
        
print(len(eyes))
print(len(faces))
for(x,y,w,h) in eyes:
     cv2.rectangle(frame_gray,(x,y),(x+w,y+h),(100,150,120),2)
for(x,y,w,h) in faces:
            cv2.rectangle(frame_gray,(x,y),(x+w,y+h),(0,255,0),2)        
plt.imshow(frame_gray)        
```
